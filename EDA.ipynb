```python
import pandas as pd
import streamlit as st
import plotly_express as px
```

```python
car_data = pd.read_csv('vehicles_us.csv')
```

```python
# Esta seccin del cdigo genera un grfico de barras
# para mostrar la frecuencia (conteo) de cada tipo de vehculo
# presente en la columna 'type' del dataset. Esto nos da
# una idea de la distribucin de las diferentes categoras de vehculos.

fig_bar = px.bar(car_data, x='type', title='Frecuencia de Tipos de Vehculos')
fig_bar.show()
```
