```python
import pandas as pd
import streamlit as st
import plotly_express as px
```

```python
car_data = pd.read_csv('vehicles_us.csv')
```

```python
# Esta sección del código genera un gráfico de barras
# para mostrar la frecuencia (conteo) de cada tipo de vehículo
# presente en la columna 'type' del dataset. Esto nos da
# una idea de la distribución de las diferentes categorías de vehículos.

fig_bar = px.bar(car_data, x='type', title='Frecuencia de Tipos de Vehículos')
fig_bar.show()
```
